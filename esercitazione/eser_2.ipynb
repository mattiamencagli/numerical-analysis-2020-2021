```python
def gen_lag_bas(q,e):
    n_q_points = len(q)
    n_e_points = e.shape[0]
    
    basis = np.ones((n_e_points,n_q_points))
    
    for i in range(n_e_points):
        for j in range(n_q_points):
            for k in range(n_q_points):
                if k!=j:
                    basis[i,j] *= (e[i]-q[k])/(q[j]-q[k])
    return basis
```

```python
n=9
N=100
q = np.linspace(0,1,n)
e = np.linspace(0,1,N)
p = gen_lag_bas(q,e)
print("p.shape = ",p.shape)
import pylab as plt
_= plt.plot(e,p) # mette dentro _ l'output atroce di matplotlib
```
